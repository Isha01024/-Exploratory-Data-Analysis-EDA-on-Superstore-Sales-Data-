```python
plt.figure(figsize=(8,6))

sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Sales'], bins=30)

plt.title("Sales Distribution")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Profit'], bins=30)

plt.title("Profit Distribution")

plt.xlabel("Profit")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Quantity'], bins=20)

plt.title("Quantity Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Quantity'], bins=20)

plt.title("Quantity Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Discount'], bins=20)

plt.title("Discount Distribution")

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar')

plt.title("Sales by Category")

plt.xlabel("Category")

plt.ylabel("Sales")

plt.show()
```

```python
sub = df.groupby('Sub-Category')['Sales'].sum().sort_values()

sub.plot(kind='barh')

plt.title("Sales by Sub Category")

plt.show()
```

```python
segment = df.groupby('Segment')['Sales'].sum()

segment.plot(kind='bar')

plt.title("Sales by Segment")

plt.show()
```

```python
region = df.groupby('Region')['Sales'].sum()

region.plot(kind='bar')

plt.title("Sales by Region")

plt.show()
```

```python
state = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

state.plot(kind='bar')

plt.title("Top 10 States")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Sales'],df['Profit'])

plt.title("Sales vs Profit")

plt.xlabel("Sales")

plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Discount'],df['Profit'])

plt.title("Discount vs Profit")

plt.xlabel("Discount")

plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Sales'])

plt.title("Sales Outliers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Profit'])

plt.title("Profit Outliers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Discount'])

plt.title("Discount Outliers")

plt.show()
```
